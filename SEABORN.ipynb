```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.drop(columns=['alone', 'alive', 'who', 'adult_male', 'embark_town', 'class'], axis=1, inplace=True)
titanic.dropna(axis=0, inplace=True)
titanic.head()
```

```python
sns.pairplot(titanic)
```

```python
sns.stripplot(data=titanic, x="fare")
```

```python
titanic.shape[1]
t = titanic.columns
t
```

```python
for i in range(titanic.shape[1]):
    c = t[i]
    sns.stripplot(data=titanic, x=c)
plt.show
```
